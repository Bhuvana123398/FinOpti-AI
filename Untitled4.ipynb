```python
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA
from scipy.optimize import minimize
from sklearn.metrics import mean_absolute_error, mean_squared_error

from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.units import inch
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet

warnings.filterwarnings("ignore", category=UserWarning)
pd.options.mode.chained_assignment = None  # avoid pandas warnings in console

# =========================
# FORECASTING
# =========================
def _auto_arima_forecast(series, periods):
    """
    Use pmdarima.auto_arima for order selection; fallback to ARIMA(1,1,1) if pmdarima not installed or fails.
    Ensures series is numeric, cleaned, and indexed properly.
    """
    series = pd.to_numeric(series, errors="coerce").dropna()
    if len(series) < 4:
        raise ValueError("Not enough historical points to forecast (need ≥ 4).")

    try:
        from pmdarima import auto_arima
        # Ensure series is with integer index or datetime index for auto_arima stability
        if series.index.dtype.kind not in ['i', 'u', 'M']:
            series.index = pd.RangeIndex(start=0, stop=len(series), step=1)

        model = auto_arima(
            series,
            start_p=0,
            start_q=0,
            max_p=3,
            max_q=3,
            max_d=2,
            seasonal=False,
            stepwise=True,
            suppress_warnings=True,
            error_action="ignore",
            trace=False,
            information_criterion='aic'
        )
        fc = model.predict(n_periods=periods)
        return pd.Series(fc)
    except Exception as e:
        # Fallback: simpler ARIMA(1,1,1)
        model = ARIMA(series, order=(1, 1, 1))
        model_fit = model.fit()
        fc = model_fit.forecast(steps=periods)
        return pd.Series(fc)

def forecast_expenditure(series: pd.Series, periods: int, expenditure_type: str = "CapEx") -> pd.Series:
    """
    Forecast expenditure using ARIMA model with auto order selection.
    Currently same method for CapEx and OpEx.
    """
    # Clean and convert series to ensure no issues
    series_clean = pd.to_numeric(series, errors="coerce").dropna()
    if len(series_clean) < 4:
        raise ValueError("Not enough historical points to forecast (need ≥ 4).")
    
    # Future: customize logic based on expenditure type if needed
    return _auto_arima_forecast(series_clean, periods)

# =========================
# OPTIMIZATION
# =========================
def optimize_expenditures(df: pd.DataFrame, inflation_rate=0.03, alpha=0.5, beta=0.5, gamma=0.001) -> pd.DataFrame:
    """
    Minimize (weighted CapEx+OpEx)*(1+inflation) + regularization toward forecasts.
    Enforces ≥80% of forecast CapEx and ≥70% of forecast OpEx.
    """
    optimized_capex, optimized_opex = [], []

    for _, row in df.iterrows():
        pred_capex, pred_opex, year = float(row['Forecast_CapEx']), float(row['Forecast_OpEx']), int(row['Year'])

        def objective(x):
            capex, opex = x
            return (alpha * capex + beta * opex) * (1 + inflation_rate) + \
                   gamma * ((capex - pred_capex) ** 2 + (opex - pred_opex) ** 2)

        constraints = [
            {"type": "ineq", "fun": lambda x, pc=pred_capex: x[0] - 0.8 * pc},
            {"type": "ineq", "fun": lambda x, po=pred_opex: x[1] - 0.7 * po},
        ]
        bounds = [(0, None), (0, None)]
        x0 = [pred_capex, pred_opex]

        result = minimize(objective, x0, method="SLSQP", bounds=bounds, constraints=constraints)

        if result.success and np.all(np.isfinite(result.x)):
            capex_opt, opex_opt = result.x
            print(f"[Optimization] Year {year}: CapEx {pred_capex:.2f} → {capex_opt:.2f}, "
                  f"OpEx {pred_opex:.2f} → {opex_opt:.2f}")
        else:
            capex_opt, opex_opt = pred_capex, pred_opex
            print(f"[Optimization Failed] Year {year} - Keeping forecasts.")

        optimized_capex.append(float(capex_opt))
        optimized_opex.append(float(opex_opt))

    out = df.copy()
    out['Optimized_CapEx'] = optimized_capex
    out['Optimized_OpEx'] = optimized_opex
    return out

# =========================
# EVALUATION
# =========================

def evaluate_forecast(bank_df: pd.DataFrame, max_year: int):
    """
    Walk-forward style: last 2-3 years as test, earlier as train.
    Uses auto_arima to select best ARIMA orders for CapEx and OpEx.
    """
    if bank_df['Year'].nunique() < 6:
        print("⚠️  Not enough history for a robust backtest (need ≥ 6 distinct years). Skipping accuracy report.")
        return {}
    
    train = bank_df[bank_df['Year'] <= max_year - 3]
    test = bank_df[bank_df['Year'] > max_year - 3]
    results = {}
    
    for expenditure_col in ['Capital Expenditure', 'Operating Expenditure']:
        series_train = train[expenditure_col].astype(float).dropna()
        series_test = test[expenditure_col].astype(float).dropna()
        
        if len(series_train) < 4:
            print(f"⚠️ Not enough training data for {expenditure_col} forecasting.")
            continue
        
        try:
            model = auto_arima(
                series_train,
                start_p=0,
                start_q=0,
                max_p=3,
                max_q=3,
                max_d=2,
                seasonal=False,
                stepwise=True,
                suppress_warnings=True,
                error_action='ignore'
            )
            
            forecast = model.predict(n_periods=len(series_test))
            forecast_series = pd.Series(forecast, index=series_test.index)
            
            mae = mean_absolute_error(series_test, forecast_series)
            rmse = np.sqrt(mean_squared_error(series_test, forecast_series))
            mape = (np.mean(np.abs((series_test.values - forecast) / np.where(series_test.values == 0, np.nan, series_test.values))) * 100)
            
            results[expenditure_col] = (mae, rmse, mape)
            print(f"\n Forecast Accuracy for {expenditure_col} (last 3 years test):")
            for metric, (mae_v, rmse_v, mape_v) in results.items():
                print(f"MAE = {mae:.2f}, RMSE = {rmse:.2f}, MAPE = {mape:.2f}%")
        
        except Exception as e:
            print(f"Forecast evaluation failed for {expenditure_col} with error: {e}")
    
    # --- PLOTS ---
    # 1) Training vs Testing Data
    plt.figure(figsize=(10, 5))
    plt.plot(train['Year'], train['Capital Expenditure'], "o-", label="Train CapEx")
    plt.plot(train['Year'], train['Operating Expenditure'], "o-", label="Train OpEx")
    plt.plot(test['Year'], test['Capital Expenditure'], "s-", label="Test CapEx")
    plt.plot(test['Year'], test['Operating Expenditure'], "s-", label="Test OpEx")
    plt.title("Train vs Test")
    plt.xlabel("Year")
    plt.ylabel("Expenditure")
    plt.legend()
    plt.grid(alpha=0.5)
    plt.show()
    
    


def evaluate_optimization(df: pd.DataFrame):
    df = df.copy()
    df['Forecast_Total'] = df['Forecast_CapEx'] + df['Forecast_OpEx']
    df['Optimized_Total'] = df['Optimized_CapEx'] + df['Optimized_OpEx']
    avg_saving = (1 - df['Optimized_Total'].sum() / df['Forecast_Total'].sum()) * 100
    print(f"\n💰 Average cost reduction from optimization: {avg_saving:.2f}%")
    return avg_saving

# =========================
# RECOMMENDATIONS
# =========================
def generate_recommendations(pred_df: pd.DataFrame, opt_df: pd.DataFrame) -> str:
    total_forecast = pred_df[['Forecast_CapEx', 'Forecast_OpEx']].sum().sum()
    total_optimized = opt_df[['Optimized_CapEx', 'Optimized_OpEx']].sum().sum()
    total_savings = total_forecast - total_optimized
    savings_pct = (total_savings / total_forecast) * 100 if total_forecast else 0.0

    capex_reduction_pct = (1 - opt_df['Optimized_CapEx'].sum() /
                           pred_df['Forecast_CapEx'].sum()) * 100 if pred_df['Forecast_CapEx'].sum() else 0.0
    opex_reduction_pct = (1 - opt_df['Optimized_OpEx'].sum() /
                          pred_df['Forecast_OpEx'].sum()) * 100 if pred_df['Forecast_OpEx'].sum() else 0.0

    recs = [
        f"Total projected savings of {total_savings:,.2f} ({savings_pct:.2f}%) achievable without disrupting critical operations.\n",
        f"CapEx can be reduced by ~{capex_reduction_pct:.2f}% via deferring non-critical upgrades and phasing multi-year rollouts.\n",
        f"OpEx can be reduced by ~{opex_reduction_pct:.2f}% via process automation and vendor renegotiations.\n",
        "Lock in supplier/service contracts ahead of renewals to mitigate inflation passthrough.\n",
        "Prioritize high-ROI initiatives; sunset low-yield spend to free budget.\n",
        "Maintain a contingency reserve for regulatory or market shocks."
    ]
    return " ".join(recs)
    
# =========================
# REPORTING (PDF + CHART)
# =========================
def plot_and_save_chart(full_df: pd.DataFrame, bank_name: str, output_dir: str) -> str:
    chart_path = os.path.join(output_dir, f"{bank_name}_chart.png")
    plt.figure(figsize=(9, 5.5))
    plt.plot(full_df["Year"], full_df["Forecast_CapEx"], "o-", label="Forecast CapEx")
    plt.plot(full_df["Year"], full_df["Optimized_CapEx"], "s-", label="Optimized CapEx")
    plt.plot(full_df["Year"], full_df["Forecast_OpEx"], "o--", label="Forecast OpEx")
    plt.plot(full_df["Year"], full_df["Optimized_OpEx"], "s--", label="Optimized OpEx")
    plt.title(f"{bank_name} - Forecast vs Optimized Expenditures")
    plt.xlabel("Year")
    plt.ylabel("Amount")
    plt.grid(alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.savefig(chart_path, bbox_inches="tight")
    plt.close()
    return chart_path

def create_pdf_report(bank_name, summary_text, full_df, recommendations, chart_path, output_path):
    pdf_path = os.path.join(output_path, f"{bank_name}_report.pdf")
    doc = SimpleDocTemplate(pdf_path, pagesize=A4)
    elements, styles = [], getSampleStyleSheet()

    elements.append(Paragraph(f"<b>{bank_name} - CapEx & OpEx Optimization Report</b>", styles['Title']))
    elements.append(Spacer(1, 0.2 * inch))
    elements.append(Paragraph("<b>Summary:</b>", styles['Heading2']))
    elements.append(Paragraph(summary_text.replace("\n", "<br/>"), styles['Normal']))
    elements.append(Spacer(1, 0.2 * inch))

    # Table
    table_data = [["Year", "Forecast CapEx", "Optimized CapEx", "Forecast OpEx", "Optimized OpEx"]]
    for _, r in full_df.iterrows():
        table_data.append([
            int(r['Year']),
            f"{float(r['Forecast_CapEx']):,.2f}",
            f"{float(r['Optimized_CapEx']):,.2f}",
            f"{float(r['Forecast_OpEx']):,.2f}",
            f"{float(r['Optimized_OpEx']):,.2f}",
        ])
    table = Table(table_data, colWidths=[1*inch]*5)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('GRID', (0, 0), (-1, -1), 0.5, colors.black),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('BOTTOMPADDING', (0,0), (-1,0), 6)
    ]))
    elements.append(table)
    elements.append(Spacer(1, 0.3 * inch))

    elements.append(Paragraph("<b>Recommendations:</b>", styles['Heading2']))
    elements.append(Paragraph(recommendations, styles['Normal']))
    elements.append(Spacer(1, 0.3 * inch))

    if chart_path and os.path.exists(chart_path):
        elements.append(Paragraph("<b>Visualization:</b>", styles['Heading2']))
        elements.append(Image(chart_path, width=5.5 * inch, height=3.5 * inch))

    doc.build(elements)
    return pdf_path

# =========================
# PIPELINE
# =========================
def process_uploaded_file(file_path, bank_name, forecast_years, output_dir, current_year):
    # Load & filter
    df = pd.read_excel(file_path)
    if 'Banks' not in df.columns:
        raise ValueError("Excel file must contain a 'Banks' column.")
    if 'Year' not in df.columns or 'Capital Expenditure' not in df.columns or 'Operating Expenditure' not in df.columns:
        raise ValueError("Excel must have 'Year', 'Capital Expenditure', and 'Operating Expenditure' columns.")

    df['Year'] = df['Year'].ffill()
    bank_df = df[df['Banks'].astype(str).str.strip().str.lower() == bank_name.strip().lower()].copy()
    if bank_df.empty:
        raise ValueError(f"No data for bank: {bank_name}")

    bank_df['Year'] = pd.to_numeric(bank_df['Year'].astype(str).str.extract(r'(\d{4})')[0], errors='coerce')
    bank_df = bank_df.dropna(subset=['Year']).astype({'Year': int})

    max_year_hist = bank_df['Year'].max()
    max_year = min(current_year - 1, max_year_hist)
    bank_df = bank_df[bank_df['Year'] <= max_year]

    if bank_df.empty or bank_df['Year'].nunique() < 4:
        raise ValueError("Not enough historical data after cleaning (need ≥ 4 years).")

    grouped = bank_df.groupby('Year', as_index=False).agg({
        'Capital Expenditure': 'sum',
        'Operating Expenditure': 'sum'
    })

    # Forecast horizon years
    start_year = max_year + 1
    forecast_years_list = list(range(start_year, start_year + forecast_years))

    # Forecasts
    #capex_forecast = forecast_expenditure(grouped['Capital Expenditure'], forecast_years).reset_index(drop=True)
    #opex_forecast = forecast_expenditure(grouped['Operating Expenditure'], forecast_years).reset_index(drop=True)
    capex_forecast = forecast_expenditure(grouped['Capital Expenditure'], forecast_years, "CapEx").reset_index(drop=True)
    opex_forecast = forecast_expenditure(grouped['Operating Expenditure'], forecast_years, "OpEx").reset_index(drop=True)
    
    pred_df = pd.DataFrame({
        'Year': forecast_years_list,
        'Forecast_CapEx': capex_forecast,
        'Forecast_OpEx': opex_forecast
    })
    print("\n Forecasted Data:\n")
    print(pred_df)

    # Optimize
    optimized_df = optimize_expenditures(pred_df)

    # Merge (for Excel, chart, and PDF)
    full_df = pred_df.merge(optimized_df[['Year', 'Optimized_CapEx', 'Optimized_OpEx']], on='Year', how='left')

    # Summary stats
    total_forecast_capex = full_df['Forecast_CapEx'].sum()
    total_forecast_opex = full_df['Forecast_OpEx'].sum()
    total_optimized_capex = full_df['Optimized_CapEx'].sum()
    total_optimized_opex = full_df['Optimized_OpEx'].sum()
    total_savings = (total_forecast_capex + total_forecast_opex) - (total_optimized_capex + total_optimized_opex)

    capex_reduction_pct = (1 - total_optimized_capex / total_forecast_capex) * 100 if total_forecast_capex else 0.0
    opex_reduction_pct = (1 - total_optimized_opex / total_forecast_opex) * 100 if total_forecast_opex else 0.0

    summary_text = (
        f"Bank: {bank_name}\n"
        f"Forecast Years: {forecast_years_list}\n\n"
        f"Total Forecast CapEx (3y): {total_forecast_capex:,.2f}\n"
        f"Total Optimized CapEx (3y): {total_optimized_capex:,.2f} "
        f"({capex_reduction_pct:.2f}% ↓)\n"
        f"Total Forecast OpEx (3y): {total_forecast_opex:,.2f}\n"
        f"Total Optimized OpEx (3y): {total_optimized_opex:,.2f} "
        f"({opex_reduction_pct:.2f}% ↓)\n"
        f"Total Estimated Savings (3y): {total_savings:,.2f}"
    )

    # Recommendations
    recommendations = generate_recommendations(pred_df, full_df)

    # Evaluations (print to console)
    evaluate_forecast(grouped.rename(columns={"Capital Expenditure":"Capital Expenditure","Operating Expenditure":"Operating Expenditure"}), max_year)
    

    evaluate_optimization(full_df)

    # Outputs
    os.makedirs(output_dir, exist_ok=True)
    excel_path = os.path.join(output_dir, f"{bank_name}_forecast_optimized.xlsx")
    full_df.to_excel(excel_path, index=False)

    chart_path = plot_and_save_chart(full_df, bank_name, output_dir)
    pdf_path = create_pdf_report(bank_name, summary_text, full_df, recommendations, chart_path, output_dir)

    return summary_text, recommendations, excel_path, chart_path, pdf_path
   

# =========================
# RUN
# =========================
if __name__ == "__main__":
    # >>> EDIT THESE PATHS/VALUES AS NEEDED <<<
    file_path = r"C:\Users\Deepika C\Final_dataset\final.xlsx"
    bank_name = input("Enter the bank name: ").strip()
    current_year = int(input("Enter Current Year (e.g., 2025): ").strip())
    forecast_years = 3
    output_dir = r"C:\Users\Deepika C\Final_dataset\outputs_test"

    print(f"\nForecasting next {forecast_years} years for {bank_name}...\n")
    summary, recs, excel_file, chart_file, pdf_file = process_uploaded_file(
        file_path, bank_name, forecast_years, output_dir, current_year
    )
    print("\n===== Summary ====="); print(summary)
    print("\n===== Recommendations ====="); print(recs)
    print(f"\nExcel: {excel_file}\nChart: {chart_file}\nPDF: {pdf_file}")
```

```python
#before opex fixing
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from scipy.optimize import minimize
from sklearn.metrics import mean_absolute_error, mean_squared_error

from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.units import inch
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet

warnings.filterwarnings("ignore", category=UserWarning)
pd.options.mode.chained_assignment = None  # avoid pandas warnings in console

# =========================
# FORECASTING
# =========================
def _auto_arima_forecast(series, periods):
    """
    Try pmdarima.auto_arima for order selection; fallback to ARIMA(1,1,1) if pmdarima not installed.
    """
    try:
        from pmdarima import auto_arima
        model = auto_arima(
            series,
            seasonal=False,
            stepwise=True,
            suppress_warnings=True,
            error_action="ignore",
            max_p=3,
            max_q=3,
            max_d=2
        )
        fc = model.predict(n_periods=periods)
        return pd.Series(fc)
    except Exception:
        # Fallback
        model = ARIMA(series, order=(1, 1, 1))
        model_fit = model.fit()
        fc = model_fit.forecast(steps=periods)
        return pd.Series(fc)

def forecast_expenditure(series: pd.Series, periods: int, expenditure_type: str = "CapEx") -> pd.Series:
    series = pd.to_numeric(series, errors="coerce").dropna()
    if len(series) < 4:
        raise ValueError("Not enough historical points to forecast (need ≥ 4).")
    
    if expenditure_type.lower() == "opex":
        # use some special forecasting logic for OpEx
        return _auto_arima_forecast(series, periods)  # or enhanced method later
    else:
        return _auto_arima_forecast(series, periods)



# =========================
# OPTIMIZATION
# =========================
def optimize_expenditures(df: pd.DataFrame, inflation_rate=0.03, alpha=0.5, beta=0.5, gamma=0.001) -> pd.DataFrame:
    """
    Minimize (weighted CapEx+OpEx)*(1+inflation) + regularization toward forecasts.
    Enforces ≥80% of forecast CapEx and ≥70% of forecast OpEx.
    """
    optimized_capex, optimized_opex = [], []

    for _, row in df.iterrows():
        pred_capex, pred_opex, year = float(row['Forecast_CapEx']), float(row['Forecast_OpEx']), int(row['Year'])

        def objective(x):
            capex, opex = x
            return (alpha * capex + beta * opex) * (1 + inflation_rate) + \
                   gamma * ((capex - pred_capex) ** 2 + (opex - pred_opex) ** 2)

        constraints = [
            {"type": "ineq", "fun": lambda x, pc=pred_capex: x[0] - 0.8 * pc},
            {"type": "ineq", "fun": lambda x, po=pred_opex: x[1] - 0.7 * po},
        ]
        bounds = [(0, None), (0, None)]
        x0 = [pred_capex, pred_opex]

        result = minimize(objective, x0, method="SLSQP", bounds=bounds, constraints=constraints)

        if result.success and np.all(np.isfinite(result.x)):
            capex_opt, opex_opt = result.x
            print(f"[Optimization] Year {year}: CapEx {pred_capex:.2f} → {capex_opt:.2f}, "
                  f"OpEx {pred_opex:.2f} → {opex_opt:.2f}")
        else:
            capex_opt, opex_opt = pred_capex, pred_opex
            print(f"[Optimization Failed] Year {year} - Keeping forecasts.")

        optimized_capex.append(float(capex_opt))
        optimized_opex.append(float(opex_opt))

    out = df.copy()
    out['Optimized_CapEx'] = optimized_capex
    out['Optimized_OpEx'] = optimized_opex
    return out

# =========================
# EVALUATION
# =========================
def evaluate_forecast(bank_df: pd.DataFrame, max_year: int):
    """
    Walk-forward style: last 2 years as test, earlier as train.
    """
    if bank_df['Year'].nunique() < 6:
        print("⚠️  Not enough history for a robust backtest (need ≥ 6 distinct years). Skipping accuracy report.")
        return {}

    train = bank_df[bank_df['Year'] <= max_year - 3]
    test = bank_df[bank_df['Year'] > max_year - 3]
    results = {}

    # CapEx
    model_capex = ARIMA(train['Capital Expenditure'], order=(1, 1, 1)).fit()
    forecast_capex = pd.Series(model_capex.forecast(steps=len(test)).values, index=test.index)
    mae = mean_absolute_error(test['Capital Expenditure'], forecast_capex)
    rmse = np.sqrt(mean_squared_error(test['Capital Expenditure'], forecast_capex))
    mape = np.mean(np.abs((test['Capital Expenditure'].values - forecast_capex.values) /
                          test['Capital Expenditure'].values)) * 100
    results['Capital Expenditure'] = (mae, rmse, mape)

    # OpEx
    model_opex = ARIMA(train['Operating Expenditure'], order=(1, 1, 1)).fit()
    forecast_opex = pd.Series(model_opex.forecast(steps=len(test)).values, index=test.index)
    mae = mean_absolute_error(test['Operating Expenditure'], forecast_opex)
    rmse = np.sqrt(mean_squared_error(test['Operating Expenditure'], forecast_opex))
    mape = np.mean(np.abs((test['Operating Expenditure'].values - forecast_opex.values) /
                          test['Operating Expenditure'].values)) * 100
    results['Operating Expenditure'] = (mae, rmse, mape)
    print("\n📊 Forecast Accuracy (last 3 years test):")
    for metric, (mae_v, rmse_v, mape_v) in results.items():
        print(f"{metric}: MAE={mae_v:.2f}, RMSE={rmse_v:.2f}, MAPE={mape_v:.2f}%")
    # --- PLOTS ---
    # 1) Training vs Testing Data
    plt.figure(figsize=(10, 5))
    plt.plot(train['Year'], train['Capital Expenditure'], "o-", label="Train CapEx")
    plt.plot(train['Year'], train['Operating Expenditure'], "o-", label="Train OpEx")
    plt.plot(test['Year'], test['Capital Expenditure'], "s-", label="Test CapEx")
    plt.plot(test['Year'], test['Operating Expenditure'], "s-", label="Test OpEx")
    plt.title("Train vs Test")
    plt.xlabel("Year")
    plt.ylabel("Expenditure")
    plt.legend()
    plt.grid(alpha=0.5)
    plt.show()
    
    return results


def evaluate_optimization(df: pd.DataFrame):
    df = df.copy()
    df['Forecast_Total'] = df['Forecast_CapEx'] + df['Forecast_OpEx']
    df['Optimized_Total'] = df['Optimized_CapEx'] + df['Optimized_OpEx']
    avg_saving = (1 - df['Optimized_Total'].sum() / df['Forecast_Total'].sum()) * 100
    print(f"\n💰 Average cost reduction from optimization: {avg_saving:.2f}%")
    return avg_saving

# =========================
# RECOMMENDATIONS
# =========================
def generate_recommendations(pred_df: pd.DataFrame, opt_df: pd.DataFrame) -> str:
    total_forecast = pred_df[['Forecast_CapEx', 'Forecast_OpEx']].sum().sum()
    total_optimized = opt_df[['Optimized_CapEx', 'Optimized_OpEx']].sum().sum()
    total_savings = total_forecast - total_optimized
    savings_pct = (total_savings / total_forecast) * 100 if total_forecast else 0.0

    capex_reduction_pct = (1 - opt_df['Optimized_CapEx'].sum() /
                           pred_df['Forecast_CapEx'].sum()) * 100 if pred_df['Forecast_CapEx'].sum() else 0.0
    opex_reduction_pct = (1 - opt_df['Optimized_OpEx'].sum() /
                          pred_df['Forecast_OpEx'].sum()) * 100 if pred_df['Forecast_OpEx'].sum() else 0.0

    recs = [
        f"Total projected savings of {total_savings:,.2f} ({savings_pct:.2f}%) achievable without disrupting critical operations.",
        f"CapEx can be reduced by ~{capex_reduction_pct:.2f}% via deferring non-critical upgrades and phasing multi-year rollouts.",
        f"OpEx can be reduced by ~{opex_reduction_pct:.2f}% via process automation and vendor renegotiations.",
        "Lock in supplier/service contracts ahead of renewals to mitigate inflation passthrough.",
        "Prioritize high-ROI initiatives; sunset low-yield spend to free budget.",
        "Maintain a contingency reserve for regulatory or market shocks."
    ]
    return " ".join(recs)
    
# =========================
# REPORTING (PDF + CHART)
# =========================
def plot_and_save_chart(full_df: pd.DataFrame, bank_name: str, output_dir: str) -> str:
    chart_path = os.path.join(output_dir, f"{bank_name}_chart.png")
    plt.figure(figsize=(9, 5.5))
    plt.plot(full_df["Year"], full_df["Forecast_CapEx"], "o-", label="Forecast CapEx")
    plt.plot(full_df["Year"], full_df["Optimized_CapEx"], "s-", label="Optimized CapEx")
    plt.plot(full_df["Year"], full_df["Forecast_OpEx"], "o--", label="Forecast OpEx")
    plt.plot(full_df["Year"], full_df["Optimized_OpEx"], "s--", label="Optimized OpEx")
    plt.title(f"{bank_name} - Forecast vs Optimized Expenditures")
    plt.xlabel("Year")
    plt.ylabel("Amount")
    plt.grid(alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.savefig(chart_path, bbox_inches="tight")
    plt.close()
    return chart_path

def create_pdf_report(bank_name, summary_text, full_df, recommendations, chart_path, output_path):
    pdf_path = os.path.join(output_path, f"{bank_name}_report.pdf")
    doc = SimpleDocTemplate(pdf_path, pagesize=A4)
    elements, styles = [], getSampleStyleSheet()

    elements.append(Paragraph(f"<b>{bank_name} - CapEx & OpEx Optimization Report</b>", styles['Title']))
    elements.append(Spacer(1, 0.2 * inch))
    elements.append(Paragraph("<b>Summary:</b>", styles['Heading2']))
    elements.append(Paragraph(summary_text.replace("\n", "<br/>"), styles['Normal']))
    elements.append(Spacer(1, 0.2 * inch))

    # Table
    table_data = [["Year", "Forecast CapEx", "Optimized CapEx", "Forecast OpEx", "Optimized OpEx"]]
    for _, r in full_df.iterrows():
        table_data.append([
            int(r['Year']),
            f"{float(r['Forecast_CapEx']):,.2f}",
            f"{float(r['Optimized_CapEx']):,.2f}",
            f"{float(r['Forecast_OpEx']):,.2f}",
            f"{float(r['Optimized_OpEx']):,.2f}",
        ])
    table = Table(table_data, colWidths=[1*inch]*5)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('GRID', (0, 0), (-1, -1), 0.5, colors.black),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('BOTTOMPADDING', (0,0), (-1,0), 6)
    ]))
    elements.append(table)
    elements.append(Spacer(1, 0.3 * inch))

    elements.append(Paragraph("<b>Recommendations:</b>", styles['Heading2']))
    elements.append(Paragraph(recommendations, styles['Normal']))
    elements.append(Spacer(1, 0.3 * inch))

    if chart_path and os.path.exists(chart_path):
        elements.append(Paragraph("<b>Visualization:</b>", styles['Heading2']))
        elements.append(Image(chart_path, width=5.5 * inch, height=3.5 * inch))

    doc.build(elements)
    return pdf_path

# =========================
# PIPELINE
# =========================
def process_uploaded_file(file_path, bank_name, forecast_years, output_dir, current_year):
    # Load & filter
    df = pd.read_excel(file_path)
    if 'Banks' not in df.columns:
        raise ValueError("Excel file must contain a 'Banks' column.")
    if 'Year' not in df.columns or 'Capital Expenditure' not in df.columns or 'Operating Expenditure' not in df.columns:
        raise ValueError("Excel must have 'Year', 'Capital Expenditure', and 'Operating Expenditure' columns.")

    df['Year'] = df['Year'].ffill()
    bank_df = df[df['Banks'].astype(str).str.strip().str.lower() == bank_name.strip().lower()].copy()
    if bank_df.empty:
        raise ValueError(f"No data for bank: {bank_name}")

    bank_df['Year'] = pd.to_numeric(bank_df['Year'].astype(str).str.extract(r'(\d{4})')[0], errors='coerce')
    bank_df = bank_df.dropna(subset=['Year']).astype({'Year': int})

    max_year_hist = bank_df['Year'].max()
    max_year = min(current_year - 1, max_year_hist)
    bank_df = bank_df[bank_df['Year'] <= max_year]

    if bank_df.empty or bank_df['Year'].nunique() < 4:
        raise ValueError("Not enough historical data after cleaning (need ≥ 4 years).")

    grouped = bank_df.groupby('Year', as_index=False).agg({
        'Capital Expenditure': 'sum',
        'Operating Expenditure': 'sum'
    })

    # Forecast horizon years
    start_year = max_year + 1
    forecast_years_list = list(range(start_year, start_year + forecast_years))

    # Forecasts
    #capex_forecast = forecast_expenditure(grouped['Capital Expenditure'], forecast_years).reset_index(drop=True)
    #opex_forecast = forecast_expenditure(grouped['Operating Expenditure'], forecast_years).reset_index(drop=True)
    capex_forecast = forecast_expenditure(grouped['Capital Expenditure'], forecast_years, "CapEx").reset_index(drop=True)
    opex_forecast = forecast_expenditure(grouped['Operating Expenditure'], forecast_years, "OpEx").reset_index(drop=True)
    
    pred_df = pd.DataFrame({
        'Year': forecast_years_list,
        'Forecast_CapEx': capex_forecast,
        'Forecast_OpEx': opex_forecast
    })

    # Optimize
    optimized_df = optimize_expenditures(pred_df)

    # Merge (for Excel, chart, and PDF)
    full_df = pred_df.merge(optimized_df[['Year', 'Optimized_CapEx', 'Optimized_OpEx']], on='Year', how='left')

    # Summary stats
    total_forecast_capex = full_df['Forecast_CapEx'].sum()
    total_forecast_opex = full_df['Forecast_OpEx'].sum()
    total_optimized_capex = full_df['Optimized_CapEx'].sum()
    total_optimized_opex = full_df['Optimized_OpEx'].sum()
    total_savings = (total_forecast_capex + total_forecast_opex) - (total_optimized_capex + total_optimized_opex)

    capex_reduction_pct = (1 - total_optimized_capex / total_forecast_capex) * 100 if total_forecast_capex else 0.0
    opex_reduction_pct = (1 - total_optimized_opex / total_forecast_opex) * 100 if total_forecast_opex else 0.0

    summary_text = (
        f"Bank: {bank_name}\n"
        f"Forecast Years: {forecast_years_list}\n\n"
        f"Total Forecast CapEx (3y): {total_forecast_capex:,.2f}\n"
        f"Total Optimized CapEx (3y): {total_optimized_capex:,.2f} "
        f"({capex_reduction_pct:.2f}% ↓)\n"
        f"Total Forecast OpEx (3y): {total_forecast_opex:,.2f}\n"
        f"Total Optimized OpEx (3y): {total_optimized_opex:,.2f} "
        f"({opex_reduction_pct:.2f}% ↓)\n"
        f"Total Estimated Savings (3y): {total_savings:,.2f}"
    )

    # Recommendations
    recommendations = generate_recommendations(pred_df, full_df)

    # Evaluations (print to console)
    evaluate_forecast(grouped.rename(columns={"Capital Expenditure":"Capital Expenditure","Operating Expenditure":"Operating Expenditure"}), max_year)
    

    evaluate_optimization(full_df)

    # Outputs
    os.makedirs(output_dir, exist_ok=True)
    excel_path = os.path.join(output_dir, f"{bank_name}_forecast_optimized.xlsx")
    full_df.to_excel(excel_path, index=False)

    chart_path = plot_and_save_chart(full_df, bank_name, output_dir)
    pdf_path = create_pdf_report(bank_name, summary_text, full_df, recommendations, chart_path, output_dir)

    return summary_text, recommendations, excel_path, chart_path, pdf_path
   

# =========================
# RUN
# =========================
if __name__ == "__main__":
    # >>> EDIT THESE PATHS/VALUES AS NEEDED <<<
    file_path = r"C:\Users\Deepika C\Final_dataset\final.xlsx"
    bank_name = input("Enter the bank name: ").strip()
    current_year = int(input("Enter Current Year (e.g., 2025): ").strip())
    forecast_years = 3
    output_dir = r"C:\Users\Deepika C\Final_dataset\outputs_test"

    print(f"\nForecasting next {forecast_years} years for {bank_name}...\n")
    summary, recs, excel_file, chart_file, pdf_file = process_uploaded_file(
        file_path, bank_name, forecast_years, output_dir, current_year
    )
    print("\n===== Summary ====="); print(summary)
    print("\n===== Recommendations ====="); print(recs)
    print(f"\nExcel: {excel_file}\nChart: {chart_file}\nPDF: {pdf_file}")
```
